### Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa
```

### Histogram

```python
bins = [40, 50, 60, 70, 80, 90, 100]

plt.hist(fifa.Overall, color = 'teal')

plt.title('Distribution of Player Skills in Fifa 2018', fontdict = {'fontweight':'bold', 'fontsize':12})
plt.xlabel('Skill Level')
plt.ylabel('Number of players')
plt.xticks(bins)

plt.savefig('Player_skill_histogram.png', dpi = 300)
plt.show()
```

### Pie Chart

```python
fifa['Preferred Foot']
```

```python
#left = fifa.loc[fifa['Preferred Foot']=='Left'].count() #This outputs an array
left = fifa.loc[fifa['Preferred Foot']=='Left'].count()[0] #This outputs the count
right = fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]


labels = ['left', 'right']
colors = ['#ababab', '#abcdef']

plt.pie([left, right], labels = labels, autopct = '%.2f %%')
plt.show()
```

```python
plt.style.use('ggplot')
plt.title('Weight Distribution of Fifa Players (in lbs)')

fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight < 125) & (fifa.Weight < 125)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >=200].count()[0]

#parsing the values directly
#plt.pie([light,light_medium,medium,medium_heavy,heavy])

#or make it an array
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = [.4, .4, 0, 0, .4]

plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance= 0.8, explode = explode)


plt.show()
```

### Box and Whiskers

```python
plt.style.use('default')

plt.figure(figsize = (6, 8))
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
juventus = fifa.loc[fifa.Club == 'Juventus']['Overall']

labels = ['Fc Barcelona', 'Real Madrid', 'Juventus']
boxes = plt.boxplot([barcelona, madrid, juventus], labels = labels, patch_artist = True, medianprops={'linewidth':2})

for box in boxes['boxes']:
    #set edge color and linewidth
    box.set(color='teal', linewidth = 2)
    #set fill color NB: parse patch_artist in the plot line when using facecolor
    box.set(facecolor='#e0e0e0')

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FiFa Overall Rating')
plt.xlabel('Soccer Teams')
# clubs = [barcelona, madrid, juventus]
# plt.boxplot(clubs)


plt.savefig('Football_comparison.png', dpi = 300)
plt.show()
```
